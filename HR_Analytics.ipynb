```python
df['experience']=df['experience'].str.replace(r'>','')
df['experience']=df['experience'].str.replace(r'<','')
df['experience']=pd.to_numeric(df['experience'])
```

```python
df=df.drop(['enrollee_id','city'],axis=1)
```

```python
df.info()
```

```python
df['gender']=df['gender'].replace({'Male':0,'Female':1,'Other':2})
df['relevent_experience']=df['relevent_experience'].replace({'Has relevent experience':0,
                                                            'No relevent experience':1})
df['enrolled_university']=df['enrolled_university'].replace({'no_enrollment':0,'Full time course':1,
                                                             'Part time course':2})
df['education_level']=df['education_level'].replace({'Graduate':0,'Masters':1,'High School':2,
                                                    'Phd':3,'Primary School':4})
df['major_discipline']=df['major_discipline'].replace({'STEM':0,'Humanities':1,'Other':2,'Business Degree':3,
                                                      'Arts':4,'No Major':5})
df['company_size']=df['company_size'].replace({'50-99':0,'100-500':1,'10000+':2,'10/49':3,'1000-4999':4,
                                              '<10':5,'500-999':6,'5000-9999':7})
df['company_type']=df['company_type'].replace({'Pvt Ltd':0,'Funded Startup':1,'Public Sector':2,
                                              'Early Stage Startup':3,'NGO':4,'Other':5})
df['last_new_job']=df['last_new_job'].replace({'1':0,'2':1,'3':2,'4':3,'>4':4,'never':5})
```

```python
df.info()
```

```python
df['gender']=df['gender'].fillna(1)
df['enrolled_university']=df['enrolled_university'].fillna(df['enrolled_university'].mean())
df['education_level']=df['education_level'].fillna(df['education_level'].mean())
df['major_discipline']=df['major_discipline'].fillna(df['major_discipline'].mean())
df['experience']=df['experience'].fillna(df['experience'].mean())
df['company_size']=df['company_size'].fillna(df['company_size'].mean())
df['company_type']=df['company_type'].fillna(df['company_type'].mean())
df['last_new_job']=df['last_new_job'].fillna(df['last_new_job'].mean())

```

```python
plt.figure(figsize=[15,10])
corr_matrix=df.corr()
sns.heatmap(corr_matrix,annot=True)
```

```python
from sklearn.model_selection import train_test_split
x=df.drop('target',axis=1)
y=df['target']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import SGDClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
rfc=RandomForestClassifier()
dtc=DecisionTreeClassifier()
lr=LogisticRegression()
knn=KNeighborsClassifier(n_neighbors=15)
sgd=SGDClassifier()
```

```python
algorithm_list=[rfc,dtc,lr,knn,sgd]
for algo in algorithm_list:
    algo.fit(x_train,y_train)
    pred=algo.predict(x_test)
```

```python
from sklearn.metrics import f1_score
score=f1_score(y_test,pred,average=None)
print(score)
```
